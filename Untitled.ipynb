```python
img = cv2.imread('lab2\\images.jfif')
```

```python
B, G, R = cv2.split(img)
```

```python
import numpy as np
gray = 0.299 * R + 0.587 * G + 0.114 * B
gray = gray.astype(np.uint8)
```

```python
cv2.imshow('Grayscale Image', gray)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
plt.imshow(gray)
```
